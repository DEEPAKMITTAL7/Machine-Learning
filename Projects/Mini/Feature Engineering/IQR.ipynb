```python
import pandas as pd
df = pd.read_csv('height_weight.csv')
df
```

```python
iq1 , iq3 = df['height'].quantile([0.25,0.75])
iq1,iq3
```

```python
iqr = iq3-iq1
iq3
```

```python
low,upp = iq1-1.5*iqr , iq3+1.5*iqr
low,upp
```

```python
df[(df['height']<low) | (df['height']>upp)]
```

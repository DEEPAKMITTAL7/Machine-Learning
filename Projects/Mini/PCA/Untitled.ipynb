```python
import pandas as pd
df = pd.read_csv('heart.csv')
df
```

```python
df['Age'].describe()
```

```python
df.describe()
```

```python
df['z_score'] = (df['Cholesterol'] - df['Cholesterol'].mean())/df['Cholesterol'].std()
df
```

```python
df1 = df[(df['z_score']<3) & (df['z_score']>-3)]
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1['Sex']=le.fit_transform(df1['Sex'])
df1['ChestPainType']=le.fit_transform(df1['ChestPainType'])
df1['RestingECG']=le.fit_transform(df1['RestingECG'])
df1['ExerciseAngina']=le.fit_transform(df1['ExerciseAngina'])
df1['ST_Slope']=le.fit_transform(df1['ST_Slope'])
```

```python
df1
```

```python
x = df1.drop(['z_score','HeartDisease'],axis='columns')
x
```

```python
y = df1.HeartDisease
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),x_scaled,y).mean()
```

```python
cross_val_score(DecisionTreeClassifier(),x_scaled,y).mean()
```

```python
cross_val_score(SVC(),x_scaled,y).mean()
```

```python
cross_val_score(RandomForestClassifier(),x_scaled,y).mean()
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
x_pca = pca.fit_transform(x_scaled)
x_pca
```

```python
cross_val_score(LogisticRegression(),x_pca,y).mean()
```

```python
cross_val_score(DecisionTreeClassifier(),x_pca,y).mean()
```

```python
cross_val_score(DecisionTreeClassifier(),x_pca,y).mean()
```

```python
cross_val_score(RandomForestClassifier(),x_pca,y).mean()
```

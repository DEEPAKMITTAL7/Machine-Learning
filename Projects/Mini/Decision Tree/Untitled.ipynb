```python
import pandas as pd
```

```python
from sklearn.datasets import load_digits
data = load_digits()
data
```

```python
data.data
```

```python
dir(data)
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
y = data.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df,data.target,test_size=0.2,random_state=0)
x_train
```

```python
from sklearn.svm import SVC
```

```python
model_linear = SVC()
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```

```python
model_rbf = SVC(kernel='rbf')
model_rbf.fit(x_train,y_train)
model_rbf.score(x_test,y_test)
```

```python
model_linear = SVC(C=2)
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```

```python
model_linear = SVC(C=2)
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```

```python
model_linear = SVC(C=10000000)
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```

```python
model_linear = SVC(gamma='auto')
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```

```python
model_rbf = SVC(kernel='rbf',C=10)
model_rbf.fit(x_train,y_train)
model_rbf.score(x_test,y_test)
```

```python
model_rbf = SVC(kernel='rbf',C=20)
model_rbf.fit(x_train,y_train)
model_rbf.score(x_test,y_test)
```

```python
model_rbf = SVC(kernel='rbf',C=1000000)
model_rbf.fit(x_train,y_train)
model_rbf.score(x_test,y_test)
```

```python
model_rbf = SVC(kernel='rbf',gamma='auto')
model_rbf.fit(x_train,y_train)
model_rbf.score(x_test,y_test)
```

```python
y = df1.Class
y
```

```python
plt.hist(y)
```

```python
#now let's check on different models
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),x,y).mean()
```

```python
cross_val_score(DecisionTreeClassifier(),x,y).mean()
```

```python
cross_val_score(SVC(),x,y).mean()
```

```python
cross_val_score(RandomForestClassifier(),x,y).mean()
```

```python
cross_val_score(GaussianNB(),x,y).mean()
```

```python
cross_val_score(KNeighborsClassifier(),x,y).mean()
```

```python
model_params ={
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    
     'decision_tree':{
        'model':DecisionTreeClassifier(),
        'params':{
            'max_features' : [None],
            'random_state' : [None],
            'max_leaf_nodes' : [None]
                     }
        },
        
    'svm': { 
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
        
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators' :[1,5,10]
        }
    },
        'knn':{
            'model':KNeighborsClassifier(),
            'params':{
                'n_neighbors' : [1,2,3]
            }
        }
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores =[]

for model_name,np in model_params.items():
    clf = GridSearchCV(np['model'],np['params'],cv=5,return_train_score=False)
    clf.fit(x,y)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
result = pd.DataFrame(scores,columns=['model','best_score','best_params'])
result
```

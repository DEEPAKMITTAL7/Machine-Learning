```python
import pandas as pd
data = pd.read_csv('bhp.csv')
data
```

```python
'''will be working on price_per_sqft'''
```

```python
mini,maxi=data['price_per_sqft'].quantile([0.001,0.999])
mini,maxi
```

```python
data[(data['price_per_sqft']>maxi) | (data['price_per_sqft']<mini)]
```

```python
df = data[(data['price_per_sqft']<maxi) & (data['price_per_sqft']>mini)]
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.stats import norm
```

```python
plt.hist(df['price_per_sqft'],density=True,bins=10,rwidth=0.8)
rng = np.arange(df['price_per_sqft'].min(),df['price_per_sqft'].max(),0.1)
plt.plot(rng,norm.pdf(rng,df['price_per_sqft'].mean(),df['price_per_sqft'].std()))
```

```python
low,upp = df['price_per_sqft'].mean()-4*df['price_per_sqft'].std(),df['price_per_sqft'].mean()+4*df['price_per_sqft'].std()
low,upp
```

```python
df1 = df[(df['price_per_sqft']>low) & (df['price_per_sqft']<upp)]
df1
```

```python
plt.hist(df1['price_per_sqft'],density=True,bins=20,rwidth=0.8)
rng = np.arange(df1['price_per_sqft'].min(),df1['price_per_sqft'].max(),0.1)
plt.plot(rng,norm.pdf(rng,df1['price_per_sqft'].mean(),df1['price_per_sqft'].std()))
```

```python
'''now using z_score'''
```

```python
df['z_score'] = (df['price_per_sqft']-df['price_per_sqft'].mean())/df['price_per_sqft'].std()
df
```

```python
df1 = df[(df['z_score']<4) & (df['z_score']>-4)]
df1
```

```python
plt.hist(df1['price_per_sqft'],density=True,bins=20,rwidth=0.8)
rng = np.arange(df1['price_per_sqft'].min(),df1['price_per_sqft'].max(),0.1)
plt.plot(rng,norm.pdf(rng,df1['price_per_sqft'].mean(),df1['price_per_sqft'].std()))
```

```python
import pandas as pd
import json
```

```python
with open('json files/t20_wc_match_results.json') as f:
    data = json.load(f)
data
```

```python
df_match_results = pd.DataFrame(data[0]['matchSummary'])
df_match_results
```

```python
df_match_results.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match_results
```

```python
df_match_results.to_csv('csv files/match_results.csv',index=False)
```

```python
with open('json files/t20_wc_batting_summary.json') as f:
    data =json.load(f)
data
```

```python
batting_summary =[]
    
for record in data:
    batting_summary.extend(record['battingSummary'])
batting_summary
```

```python
df_batting_summary = pd.DataFrame(batting_summary)
df_batting_summary
```

```python
for i in range(len(df_batting_summary)):
    if len(df_batting_summary['dismissal'][i])>0:
        df_batting_summary['dismissal'][i]='out'
    else:
        df_batting_summary['dismissal'][i]='not_out'
df_batting_summary

#alternative using lambda function
#df_batting_summary['dismissal'] = df_batting_summary['dismissal'].apply(lambda x : 'out' if len(x)>0 else 'not-out' ) 
     
```

```python
df_batting_summary.head(15)
```

```python
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('',''))
df_batting_summary.head(15)
```

```python
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting_summary
```

```python
df_match_results[(df_match_results['team1']=='England') | (df_match_results['team2']=='England')]
```

```python
match_ids_dict ={}

for index,row in df_match_results.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_ids_dict[key1]=row['match_id']
    match_ids_dict[key2]=row['match_id']

match_ids_dict
```

```python
df_batting_summary['match_id'] = df_batting_summary['match'].map(match_ids_dict)
df_batting_summary
```

```python
df_batting_summary.to_csv('csv files/batting_summary.csv',index=False)
```

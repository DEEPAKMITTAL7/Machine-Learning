```python
'''
I have decide to keep a range and store a avg of it
 and going to drop rest for now'''
```

```python
def avg(x):
    t = x.split('-')
    if len(t)==2:
        return (float(t[0])+float(t[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
avg('2100')
```

```python
avg('2100 - 2500 ')
```

```python
avg('3400sqft')
```

```python
data['total_sqft'] = data['total_sqft'].apply(avg)
data
```

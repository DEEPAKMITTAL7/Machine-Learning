```python
import pandas as pd
```

```python
'''
This is also classification 

Why DECISION TREE ????

In logistic regression, we have data seprated very well, but in real life it's not.

We need to build a tree like structure to understand and work on data.

sklearn is super powerful library once again.


''' 
```

```python
data = pd.read_csv('salaries.csv')
data
```

```python
input = data.drop('salary_more_then_100k',axis='columns')
y = data[['salary_more_then_100k']]
```

```python
input
```

```python
y
```

```python
'''
Our all inputs are words i.e non-numeric lets call LabelEncoder to help us
'''
```

```python
from sklearn.preprocessing import LabelEncoder
le_company =  LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['company_n']  = le_company.fit_transform(input['company'])
input['job_n']  = le_job.fit_transform(input['job'])
input['degree_n']  = le_degree.fit_transform(input['degree'])
input
```

```python
x  = input.drop(['company','job','degree'],axis='columns')
x
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[2,0,1]])
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
x,y
```

```python
def gradient_descent(x,y):
    '''we have y = mx+c line
    mse = (1/n)*sum((y-ypredicted)**2)
    and we are tring to get mapping of x and y
    
    requiremenst 
    1. Cost Function of mderivative and cderivate
    2. Learning rate 
    3. Learningmatrix of m and c 
    '''
    m,c=0,0
    learning_rate = 0.08
    
    n = len(x) #assuming len(x)==len(y)
    
    for i in range(10000):
        y_predicted = m * x + c
        cost = (1/n) * sum( (y-y_predicted)**2 )
        md = -(2/n) * sum(x*(y-y_predicted))
        cd = -(2/n) * sum(y-y_predicted)
        m = m - learning_rate * md
        c = c - learning_rate * cd
        print("m {},c {}, iteraion {}, cost {}".format(m,c,i,cost))

gradient_descent(x,y)
```

```python
'''

underfitting -> where most of the points are lying out the the line
match = m0 + m1*age1 

overfitting -> where all the points are on the line
match = m0 + m1*age1 + m2*age2 + m3*age3 + m4*age4

balanced fit -> where most of the points are following the line
match = m0 + m1*age1 + m2*age2

'''
```

```python
'''
Our task is make overfitting into balanced fit i.e making m3 and m4 close to zero

'''
```

```python
import pandas as pd
```

```python
data = pd.read_csv("Melbourne_housing_FULL.csv")
data
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data.head()
```

```python
data.nunique()
```

```python
data.shape
```

```python
'''
discarding some columns
'''
```

```python
data.drop(['Address','Postcode','Date','YearBuilt','Lattitude','Longtitude'],axis='columns',inplace=True)
data
```

```python
data.shape
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data[['Propertycount','Distance','Bedroom2','Car','Bathroom']] = data[['Propertycount','Distance','Bedroom2','Car','Bathroom']].fillna(0)
data
```

```python
data.isna().sum()
```

```python
data['Landsize'] = data['Landsize'].fillna(data['Landsize'].mean())
data['BuildingArea'] = data['BuildingArea'].fillna(data['BuildingArea'].mean())
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data,drop_first=True)
data
```

```python
x = data.drop(['Price'],axis=1)
x
```

```python
y = data['Price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
'''
Have you noticed what happedned ????

Our model performed well over trained than test... this means our model is overfitted

Let's call L1 Regularization i.e Lasso for help

'''
```

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_model.fit(x_train,y_train)
```

```python
lasso_model.score(x_test,y_test)
```

```python
lasso_model.score(x_train,y_train)
```

```python
'''
Score improved a lot by margin

L2 Regulariztion i.e Ridge

'''
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```

```python
ridge_model.score(x_train,y_train)
```

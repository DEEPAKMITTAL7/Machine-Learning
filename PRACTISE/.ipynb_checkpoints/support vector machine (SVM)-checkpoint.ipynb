```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df['target'] = data.target
df
```

```python
data.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: data.target_names[x])
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='r',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='y',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='g',marker='*')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='r',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='y',marker='*')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='g',marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'],axis='columns')
x
```

```python
y = df[['target']]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
